```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Read data from folders
data_df = pd.read_csv("cleaned_covid_data.csv")
data_df.head()
```

```python
#Which countries were most affected by Covid-19
#group by location getting max cases and fetching the top 25
most_cases_df = data_df.groupby(by = ["location"]).max("total_cases").nlargest(25, "total_cases")
most_cases_df.head(25)
```

```python
#Which countries were lease affected by Covid-19
#group by location getting max cases and fetching the last 25
least_cases_df = data_df.groupby(by = ["location"]).max("total_cases").nsmallest(25, "total_cases")
least_cases_df.head(25)
```

```python
# What income group the countries affected fall into?
ax = most_cases_df["gdp_per_capita"].plot(label = "Most")
least_cases_df["gdp_per_capita"].plot(ax = ax, label = "Least")
plt.legend()
plt.xticks([])
```

```python
#Is GDP per capita a determinant of cross-country differentials in impact of Covid-19?
groups = data_df.groupby("location")
keys = groups.groups.keys()
gdp_correlation = []
for i in keys:
    gdp_correlation.append(( i , groups.get_group(i)["gdp_per_capita"].corr(groups.get_group(i)["stringency_index"]),groups.get_group(i)["gdp_per_capita"].corr(groups.get_group(i)["population"]),groups.get_group(i)["gdp_per_capita"].corr(groups.get_group(i)["HDI"])))

gdp_correlation_df = pd.DataFrame(gdp_correlation, columns=["country", "stringency_corr","population_corr","HDI_corr"])
gdp_correlation_df.head()
```

```python
#Is HDI a determinant of cross-country differentials in impact of Covid-19?
groups = data_df.groupby("location")
keys = groups.groups.keys()
hdi_correlation = []
for i in keys:
    hdi_correlation.append(( i , groups.get_group(i)["HDI"].corr(groups.get_group(i)["stringency_index"]),groups.get_group(i)["HDI"].corr(groups.get_group(i)["population"]),groups.get_group(i)["HDI"].corr(groups.get_group(i)["gdp_per_capita"])))

hdi_correlation_df = pd.DataFrame(hdi_correlation, columns=["country", "stringency_corr","population_corr","gdp_corr"])
hdi_correlation_df.head()
```
